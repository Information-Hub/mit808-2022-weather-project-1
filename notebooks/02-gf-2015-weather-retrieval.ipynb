```python
import os
import pandas as pd
```

Iteration code

```python
# assign directory
directory = "C:/Users/gf2704928/Documents/_MIT (Desktop)/__DataFor808/RawData/2008_2015temperature/"
Cumulative2015 = pd.DataFrame(columns = ['clim_no','record_date','tmax','tmin'])
 
# iterate over files in that directory
for filename in os.listdir(directory):
    fileContent = pd.read_csv("C:/Users/gf2704928/Documents/_MIT (Desktop)/__DataFor808/RawData/2008_2015temperature/" + filename, header = None)
    
    for format in ["%d/%m/%Y","%Y/%m/%d"]:
        try:
            fileContent[1] = pd.to_datetime(fileContent[1], format= format)
        except:
            pass

    fileContent2015 = fileContent[(fileContent[1].dt.year == 2015) & (fileContent[1].dt.month < 6)]
    fileContent2015 = fileContent2015.drop(columns = [3,5])
    fileContent2015 = fileContent2015.rename(columns = {0:"clim_no",1:"record_date",2:"tmax",4:"tmin"})
    Cumulative2015 = Cumulative2015.append(fileContent2015)
    
print(Cumulative2015.head())
Cumulative2015.to_csv('C:/Users/gf2704928/Documents/_MIT (Desktop)/__DataFor808/Modeling/Cumulative2015temp.csv')
```
